### 명령어 및 함수 정리

#### 리스트

* 다차원 리스트는 대괄호 안에 대괄호를 넣어 표현한다.
    * [[1,2,3] [4,5,6]]
* 리스트에 저장된 데이터 개수 : `len(listname)`
    * `len(a)`
* 리스트의 마지막에 데이터 추가 : `listname.append(data)`
    * `a.append(1)`
* 리스트의 원하는 인덱스에 데이터 삽입 : `listname.insert(index, data)`
    * `a.insert(1,2)`
    * 데이터 개수가 n개인 리스트에서 n+2 인덱스에 데이터를 삽입할 수는 없다.
* 리스트에서 **첫번째로** 나오는 데이터 삭제 : `listname.remove(data)`
    * `a.remove(1)`
* 오름차순 정렬 : `listname.sort()`
    * `a.sort()`
* 역순배열 : `listname.reverse()`
    * `a.reverse()`

### 그래프 시각화 : `matplotlib`

#### 기본 설정
* 패키지 불러오기
    `import matplotlib.pyplot as plot`
* 그래프 제목 설정
    `plt.title("제목", fontsize=16)`
* 축 이름 설정
    `plt.xlabel("x축제목", fontsize=12)`
    `plt.ylabel("y축제목", fontsize=12)`

#### 그래프 종류
* 산점도 그래프
    * 2개의 연속형 변수 간의 관계를 보기 위하여 직교 자표의 평면에 관측점을 찍어 만든 통계 그래프
    * `plt.scatter(x데이터, y데이터)`

* 라인 그래프
    * 산점도의 각 점을 이은 그래프
    * `plt.plot(x데이터, y데이터)`
